### Missing Values Before Imputation

To see the original missing values, we'll load the raw dataset again into a new DataFrame and check its null counts.

```python
import pandas as pd

df_raw = pd.read_csv('data/urban_street_food_vendor_survival_dataset.csv')
print("Missing values in the raw dataset before imputation:")
print(df_raw.isnull().sum()[df_raw.isnull().sum() > 0])
```

```python
# Impute numerical columns with median or mean
df['vendor_age_years'] = df['vendor_age_years'].fillna(df['vendor_age_years'].median())
df['years_in_business'] = df['years_in_business'].fillna(df['years_in_business'].median())
df['avg_daily_revenue_inr'] = df['avg_daily_revenue_inr'].fillna(df['avg_daily_revenue_inr'].median())
df['avg_daily_customers'] = df['avg_daily_customers'].fillna(df['avg_daily_customers'].median())
df['monthly_stall_rent_inr'] = df['monthly_stall_rent_inr'].fillna(df['monthly_stall_rent_inr'].median())
df['hours_open_per_day'] = df['hours_open_per_day'].fillna(df['hours_open_per_day'].mean())
df['competition_within_100m'] = df['competition_within_100m'].fillna(df['competition_within_100m'].median())
df['monthly_health_inspection_score'] = df['monthly_health_inspection_score'].fillna(df['monthly_health_inspection_score'].mean())
df['customer_complaint_rate'] = df['customer_complaint_rate'].fillna(df['customer_complaint_rate'].median())

# Impute binary/boolean columns with mode
df['had_fine_last_year'] = df['had_fine_last_year'].fillna(df['had_fine_last_year'].mode()[0])
df['has_online_presence'] = df['has_online_presence'].fillna(df['has_online_presence'].mode()[0])

# Impute categorical column 'license_status' with mode
df['license_status'] = df['license_status'].fillna(df['license_status'].mode()[0])

# Convert binary/boolean columns to int if they aren't already
df['had_fine_last_year'] = df['had_fine_last_year'].astype(int)
df['has_online_presence'] = df['has_online_presence'].astype(int)
df['vendor_survived'] = df['vendor_survived'].astype(int)

print("Missing values after consolidated imputation and type conversion:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Drop 'vendor_id' as it is an identifier and not suitable for correlation
numerical_df = numerical_df.drop(columns=['vendor_id'], errors='ignore')

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plotting the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

### Interpreting the Correlation Heatmap

The heatmap displays the Pearson correlation coefficients between all pairs of numerical features in the dataset. Here's how to interpret it:

*   **Color Scale (cmap='coolwarm')**: The color bar on the right indicates the strength and direction of the correlation:
    *   **Red/Darker Colors**: Indicate a strong positive correlation (closer to +1). This means that as one variable increases, the other tends to increase as well.
    *   **Blue/Lighter Colors**: Indicate a strong negative correlation (closer to -1). This means that as one variable increases, the other tends to decrease.
    *   **White/Neutral Colors**: Indicate a weak or no linear correlation (closer to 0).

*   **Values (annot=True)**: The numbers within each cell are the correlation coefficients themselves.

*   **Diagonal**: The diagonal elements are always 1, representing the correlation of a variable with itself.

**What makes a heatmap 'good' or 'bad' depends on your analysis goals:**

*   **Good for Feature Selection (Predictive Modeling)**:
    *   **High correlation with target variable (`vendor_survived`)**: Features that show a strong positive or negative correlation with the target variable are generally good candidates for predictive models, as they likely have significant influence on the outcome.
    *   **Low correlation between independent features**: High correlation between two independent (predictor) features can indicate multicollinearity, which might negatively impact some models (e.g., linear regression). If you see very dark red or blue squares between predictor variables, you might consider removing one of them or using dimensionality reduction techniques.

*   **Good for Understanding Relationships (Exploratory Data Analysis)**:
    *   Clear patterns of strong positive or negative correlations can reveal underlying relationships in your data that might not be immediately obvious. For example, if `vendor_age_years` and `years_in_business` are highly correlated, it suggests older vendors tend to have more experience, which is an expected and interpretable relationship.

In summary, this heatmap is a valuable tool for understanding the dataset's structure, identifying potential multicollinearity, and pinpointing features that may be most relevant for predicting vendor survival.

### Feature Selection based on Correlation Heatmap

Based on the observed correlation values from the heatmap, several features show very low absolute correlation with the target variable `vendor_survived`. These features might not contribute significantly to predicting vendor survival and can be removed to simplify the model and reduce dimensionality.

```python
# Features identified for removal due to very low absolute correlation with 'vendor_survived' (< 0.01)
features_to_drop = [
    'vendor_age_years',
    'years_in_business',
    'monthly_stall_rent_inr',
    'num_helpers',
    'hours_open_per_day',
    'competition_within_100m',
    'monthly_health_inspection_score',
    'had_fine_last_year'
]

print(f"Original DataFrame shape: {df.shape}")
df = df.drop(columns=features_to_drop, errors='ignore')
print(f"DataFrame shape after dropping features: {df.shape}")
print(f"Removed features: {features_to_drop}")

# Display the first few rows of the updated DataFrame
display(df.head())
```

### New Correlation Heatmap After Feature Removal

Let's visualize the correlation matrix again with the reduced set of features to confirm the changes and understand the relationships among the remaining variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation calculation from the updated df
numerical_df_cleaned = df.select_dtypes(include=['float64', 'int64'])

# Drop 'vendor_id' as it is an identifier and not suitable for correlation
numerical_df_cleaned = numerical_df_cleaned.drop(columns=['vendor_id'], errors='ignore')

# Calculate the correlation matrix for the cleaned numerical features
correlation_matrix_cleaned = numerical_df_cleaned.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_cleaned, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Remaining Numerical Features (After Removal)')
plt.show()
```
